```python
import random

# ê²Œì„ì˜ ê¸°ë³¸ì´ ë˜ëŠ” Character í´ë˜ìŠ¤
class Character:
    def __init__(self, name, gender, job, level, hp, power, defense, mp=0):
        self.name = name            # ìºë¦­í„° ì´ë¦„
        self.gender = gender        # ìºë¦­í„° ì„±ë³„
        self.job = job             # ìºë¦­í„° ì§ì—…
        self.level = level          # ë ˆë²¨
        self.hp = hp               # í˜„ì¬ ì²´ë ¥
        self.max_hp = hp           # ìµœëŒ€ ì²´ë ¥
        self.power = power         # ê³µê²©ë ¥
        self.defense = defense     # ë°©ì–´ë ¥
        self.mp = mp               # í˜„ì¬ ë§ˆë‚˜
        self.max_mp = mp           # ìµœëŒ€ ë§ˆë‚˜

    # ìºë¦­í„°ê°€ ì‚´ì•„ìˆëŠ”ì§€ í™•ì¸í•˜ëŠ” ë©”ì„œë“œ
    def is_alive(self):
        return self.hp > 0

    # ë°ë¯¸ì§€ë¥¼ ë°›ì•˜ì„ ë•Œ ì²˜ë¦¬í•˜ëŠ” ë©”ì„œë“œ
    def take_damage(self, damage):
        if self.defense >= damage:  # ë°©ì–´ë ¥ì´ ë°ë¯¸ì§€ë³´ë‹¤ ë†’ìœ¼ë©´ ë°ë¯¸ì§€ë¥¼ ë°›ì§€ ì•ŠìŒ
            return
        real_damage = damage - self.defense # ì‹¤ì œë¡œ ë°›ëŠ” ë°ë¯¸ì§€ = ê³µê²©í•œ ë°ë¯¸ì§€ - ë‚´ ë°©ì–´ë ¥
        self.hp -= real_damage # í˜„ì¬ ì²´ë ¥ì—ì„œ ë°ë¯¸ì§€ë§Œí¼ ê¹ê¸°
        print(f"{self.name}ì´(ê°€) {real_damage}ì˜ ê³µê²©ì„ ë‹¹í–ˆë‹¤!!!")
        print(f"ë‚¨ì€ ì²´ë ¥: {self.hp}")

    # ê³µê²© ë©”ì„œë“œ
    def attack_target(self, target):
        damage = random.randint(1, self.power) # 1ë¶€í„° ë‚´ ê³µê²©ë ¥ ì‚¬ì´ì—ì„œ ëœë¤í•˜ê²Œ ë°ë¯¸ì§€ ì •í•˜ê¸°
        print(f"{self.name}ì´(ê°€) {target.name}ì„(ë¥¼) ì‚¬ì •ì—†ì´ íŒ¹ë‹ˆë‹¤!!!")
        target.take_damage(damage) # ìƒëŒ€í•œí…Œ ë°ë¯¸ì§€ ì£¼ê¸°

# Characterë¥¼ ìƒì†ë°›ëŠ” Player í´ë˜ìŠ¤
class Player(Character):
    def __init__(self, name, gender, job):
        # ì§ì—…ë³„ ì´ˆê¸° ìŠ¤íƒ¯ ì„¤ì •
        if job == "ì „ì‚¬": 
            super().__init__(name, gender, job, 1, 150, 30, 10)  # ì²´ë ¥ê³¼ ë°©ì–´ë ¥ì´ ë†’ìŒ
        elif job == "ë„ì ":
            super().__init__(name, gender, job, 1, 100, 35, 5)   # ê³µê²©ë ¥ì´ ë†’ìŒ
        elif job == "ë§ˆë²•ì‚¬":
            super().__init__(name, gender, job, 1, 80, 40, 3, 100)  # ê³µê²©ë ¥ì´ ë§¤ìš° ë†’ì§€ë§Œ ì²´ë ¥ì´ ë‚®ìŒ
        elif job == "ê¶ìˆ˜":
            super().__init__(name, gender, job, 1, 90, 35, 4)    # ë°¸ëŸ°ìŠ¤í˜•
        
        self.exp = 0               # ê²½í—˜ì¹˜
        self.attack_count = 0      # ê³µê²© íšŸìˆ˜ (í•„ì‚´ê¸°ìš©)
        self.stat_points = 0       # ì‚¬ìš© ê°€ëŠ¥í•œ ìŠ¤íƒ¯ í¬ì¸íŠ¸

    # ìºë¦­í„° ìƒíƒœë¥¼ ë³´ì—¬ì£¼ëŠ” ë©”ì„œë“œ
    def show_status(self):
        print(f"\n{'='*20} ìºë¦­í„° ì •ë³´ {'='*20}")
        print(f"ì´ë¦„: {self.name}")
        print(f"ì„±ë³„: {self.gender}")
        print(f"ì§ì—…: {self.job}")
        print(f"ë ˆë²¨: {self.level}")
        print(f"ì²´ë ¥: {self.hp}/{self.max_hp}")
        if self.mp > 0:  # ë§ˆë²•ì‚¬ì˜ ê²½ìš°ì—ë§Œ MP í‘œì‹œ
            print(f"ë§ˆë‚˜: {self.mp}/{self.max_mp}")
        print(f"ê³µê²©ë ¥: {self.power}")
        print(f"ë°©ì–´ë ¥: {self.defense}")
        print(f"ê²½í—˜ì¹˜: {self.exp}/50")
        print(f"ë‚¨ì€ ìŠ¤íƒ¯ í¬ì¸íŠ¸: {self.stat_points}")
        print('='*50)

    # ë ˆë²¨ì—… ì²˜ë¦¬ ë©”ì„œë“œ
    def level_up(self):
        self.level += 1
        self.stat_points += 3  # ë ˆë²¨ì—… ë‹¹ 3í¬ì¸íŠ¸ ì§€ê¸‰
        
        # ì§ì—…ë³„ ê¸°ë³¸ ë ˆë²¨ì—… ë³´ë„ˆìŠ¤
        if self.job == "ì „ì‚¬":
            self.max_hp += 20
            self.power += 5
            self.defense += 4
        elif self.job == "ë„ì ":
            self.max_hp += 15
            self.power += 6
            self.defense += 3
        elif self.job == "ë§ˆë²•ì‚¬":
            self.max_hp += 10
            self.power += 8
            self.defense += 2
            self.max_mp += 20
        elif self.job == "ê¶ìˆ˜":
            self.max_hp += 12
            self.power += 7
            self.defense += 3
            
        # ë ˆë²¨ì—… ì‹œ ì²´ë ¥ê³¼ ë§ˆë‚˜ íšŒë³µ
        self.hp = self.max_hp
        if hasattr(self, 'max_mp'):
            self.mp = self.max_mp
        self.exp = 0
        
        print(f"\nğŸ‰ ë ˆë²¨ ì—…! í˜„ì¬ ë ˆë²¨: {self.level}")
        print(f"ìŠ¤íƒ¯ í¬ì¸íŠ¸ 3ì ì„ íšë“í–ˆìŠµë‹ˆë‹¤. (í˜„ì¬ ë³´ìœ  í¬ì¸íŠ¸: {self.stat_points})")
        self.show_status()
        self.distribute_stat_points()

    # ìŠ¤íƒ¯ í¬ì¸íŠ¸ ë¶„ë°° ë©”ì„œë“œ
    def distribute_stat_points(self):
        while self.stat_points > 0:
            print(f"\nğŸ“Š ì‚¬ìš© ê°€ëŠ¥í•œ ìŠ¤íƒ¯ í¬ì¸íŠ¸: {self.stat_points}")
            print("\nê°•í™”í•  ëŠ¥ë ¥ì¹˜ë¥¼ ì„ íƒí•˜ì„¸ìš”:")
            print("1. â¤ï¸ ì²´ë ¥ (+20)")
            print("2. âš”ï¸ ê³µê²©ë ¥ (+3)")
            print("3. ğŸ›¡ï¸ ë°©ì–´ë ¥ (+2)")
            if self.job == "ë§ˆë²•ì‚¬":
                print("4. ğŸŒŸ ë§ˆë‚˜ (+15)")
            print("0. ğŸ’¾ ëŠ¥ë ¥ì¹˜ ì €ì¥í•˜ê¸°")
            
            try:
                choice = input("\nì„ íƒ (ìˆ«ì ì…ë ¥): ").strip()
                
                if choice == "0":
                    print("\nìŠ¤íƒ¯ ë¶„ë°°ë¥¼ ì¢…ë£Œí•©ë‹ˆë‹¤.")
                    break
                    
                if not choice.isdigit() or int(choice) < 1 or int(choice) > (4 if self.job == "ë§ˆë²•ì‚¬" else 3):
                    print("\nâŒ ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”")
                    continue
                
                choice = int(choice)
                
                if choice == 1:
                    self.max_hp += 20
                    self.hp += 20
                    print("\nì²´ë ¥ì´ 20 ì¦ê°€í–ˆìŠµë‹ˆë‹¤")
                elif choice == 2:
                    self.power += 3
                    print("\nê³µê²©ë ¥ì´ 3 ì¦ê°€í–ˆìŠµë‹ˆë‹¤")
                elif choice == 3:
                    self.defense += 2
                    print("\në°©ì–´ë ¥ì´ 2 ì¦ê°€í–ˆìŠµë‹ˆë‹¤")
                elif choice == 4 and self.job == "ë§ˆë²•ì‚¬":
                    self.max_mp += 15
                    self.mp += 15
                    print("\në§ˆë‚˜ê°€ 15 ì¦ê°€í–ˆìŠµë‹ˆë‹¤")
                
                self.stat_points -= 1
                self.show_status()
                
            except ValueError:
                print("\nâŒ ì˜¬ë°”ë¥¸ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”")

    # ê²½í—˜ì¹˜ íšë“ ë©”ì„œë“œ
    def gain_experience(self, exp):
        self.exp += exp
        print(f"ê²½í—˜ì¹˜ê°€ {exp} ì˜¬ëìŠµë‹ˆë‹¤")
        if self.exp >= 50:  # 50 ê²½í—˜ì¹˜ë¥¼ ë‹¬ì„±í•˜ë©´ ë ˆë²¨ì—…
            self.level_up()

    # í•„ì‚´ê¸° ê³µê²© ë©”ì„œë“œ
    def special_attack(self, target):
        damage = self.power * 5  # ê¸°ë³¸ ê³µê²©ë ¥ì˜ 5ë°°
        print(f"ğŸ˜ ì¡°ê±´ì´ ì¶©ì¡±ë˜ì–´ í•„ì‚´ê¸°ë¥¼ ì‚¬ìš©í•©ë‹ˆë‹¤!!!!")
        print(f"ğŸ’¥ {self.name}ì´(ê°€) {target.name}ì—ê²Œ í‘ì—¼ë£¡ì„ ì‚¬ìš©í•©ë‹ˆë‹¤!!!")
        target.take_damage(damage)
        self.attack_count = 0  # ê³µê²© íšŸìˆ˜ ì´ˆê¸°í™”

    # ê¸°ë³¸ ê³µê²© ë©”ì„œë“œ (í•„ì‚´ê¸° í¬í•¨)
    def attack_target(self, target):
        self.attack_count += 1  # ê³µê²© íšŸìˆ˜ ì¦ê°€
        
        # 5ë²ˆì§¸ ê³µê²©ì´ë©´ í•„ì‚´ê¸° ì‚¬ìš©
        if self.attack_count >= 5:
            self.special_attack(target)
        else:
            super().attack_target(target)

# Characterë¥¼ ìƒì†ë°›ëŠ” Monster í´ë˜ìŠ¤
class Monster(Character):
    def __init__(self, name, level):
        hp = random.randint(10, 30) * level
        power = random.randint(5, 20) * level
        defense = random.randint(1, 5) * level
        super().__init__(name, level, "Monster", level, hp, power, defense)

# Monsterë¥¼ ìƒì†ë°›ëŠ” BossMonster í´ë˜ìŠ¤
class BossMonster(Monster):
    def __init__(self, name, level, base_hp, base_power):
        self.name = name
        self.level = level
        self.hp = base_hp * 2       # ê¸°ë³¸ ì²´ë ¥ì˜ 2ë°°
        self.max_hp = self.hp
        self.power = base_power * 2  # ê¸°ë³¸ ê³µê²©ë ¥ì˜ 2ë°°
        self.defense = random.randint(1, 5) * level

# ë³´ìŠ¤ ëª¬ìŠ¤í„° ë“±ì¥ ì‹œ ì„ íƒì§€ ì²˜ë¦¬ í•¨ìˆ˜
def boss_encounter(player):
    while True:
        print("\nğŸ”¥ ë³´ìŠ¤ëª¬ìŠ¤í„° ëŒ€ì™• ê³ ë¸”ë¦° ë‘ë‘¥ ë“±ì¥ ğŸ”¥")
        print("\nì–´ë–»ê²Œ í•˜ì‹œê² ìŠµë‹ˆê¹Œ?")
        print("1. âš”ï¸ ì‹¸ìš´ë‹¤")
        print("2. ğŸƒ ë„ë§ê°„ë‹¤")
        
        choice = input("\nì„ íƒí•´ì£¼ì„¸ìš” (1 ë˜ëŠ” 2): ").strip()
        
        if choice == "1":
            print("\nğŸ—¡ï¸ ì „íˆ¬ë¥¼ ì‹œì‘í•©ë‹ˆë‹¤!!!!")
            return True
        elif choice == "2":
            print("\nğŸƒâ€â™‚ï¸ ì „íˆ¬ì—ì„œ ë„ë§ì³¤ìŠµë‹ˆë‹¤...")
            print("ì‚´ì•˜ì§€ë§Œ... ë§ˆì„ ì‚¬ëŒë“¤ì´ ì†ê°€ë½ì§ˆ í•©ë‹ˆë‹¤.")
            return False
        else:
            print("\nâŒ 1 ë˜ëŠ” 2ë§Œ ì…ë ¥!!!!!!!!!!!")

# ì „íˆ¬ ì²˜ë¦¬ í•¨ìˆ˜
def battle(player, monster):
    print(f"\nâš”ï¸ {monster.name}ê³¼(ì™€)ì˜ ì „íˆ¬ê°€ ì‹œì‘ë©ë‹ˆë‹¤!!!!!!")
    print(f"{monster.name}ì˜ ì •ë³´ - ì²´ë ¥: {monster.hp}, ê³µê²©ë ¥: {monster.power}")
    
    while player.is_alive() and monster.is_alive():
        player.attack_target(monster)
        if monster.is_alive():
            monster.attack_target(player)
            
    if player.is_alive():
        exp = monster.level * 3
        player.gain_experience(exp)
        print(f'ğŸ‰ ì´ê²¼ë“œì•„!!!!!!')
        return True
    else:
        print(f'ğŸ˜¢ ë‚˜ì•½í•˜êµ¬ë‚˜ ì¸ê°„')
        return False

# ìºë¦­í„° ìƒì„± í•¨ìˆ˜
def create_character():
    print("\nğŸ® ìºë¦­í„° ìƒì„±ì„ ì‹œì‘í•©ë‹ˆë‹¤!")
    
    # ì´ë¦„ ì…ë ¥
    while True:
        name = input("\nğŸ‘¤ ë‹¹ì‹ ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ").strip()
        if name:  # ì´ë¦„ì´ ë¹„ì–´ìˆì§€ ì•Šìœ¼ë©´
            break
        print("ì´ë¦„ì„ ë°˜ë“œì‹œ ì…ë ¥í•´ì•¼ í•©ë‹ˆë‹¤!")
    
    # ì„±ë³„ ì„ íƒ - ìˆ˜ì •ëœ ë¶€ë¶„
    while True:
        print("\nğŸ‘¥ ì„±ë³„ì„ ì„ íƒí•˜ì„¸ìš”:")
        print("1. ë‚¨ì„±")
        print("2. ì—¬ì„±")
        gender_choice = input("ì„ íƒ (1 ë˜ëŠ” 2): ").strip()
        
        if gender_choice == "1":
            gender = "ë‚¨"
            break
        elif gender_choice == "2":
            gender = "ì—¬"
            break
        print("1 ë˜ëŠ” 2ë§Œ ì…ë ¥ ê°€ëŠ¥í•©ë‹ˆë‹¤!")
    
    # ì§ì—… ì„ íƒ
    jobs = ['ì „ì‚¬', 'ë„ì ', 'ë§ˆë²•ì‚¬', 'ê¶ìˆ˜']
    print("\nâš”ï¸ ì§ì—… ëª©ë¡:")
    for i, job in enumerate(jobs, 1):
        print(f"{i}. {job}")
    
    while True:
        try:
            choice = int(input("\nì§ì—…ì„ ì„ íƒí•˜ì„¸ìš” (ë²ˆí˜¸ ì…ë ¥): "))
            if 1 <= choice <= len(jobs):
                job = jobs[choice-1]
                break
            print(f"1ë¶€í„° {len(jobs)}ê¹Œì§€ì˜ ìˆ«ìë§Œ ì…ë ¥ ê°€ëŠ¥í•©ë‹ˆë‹¤")
        except ValueError:
            print("ìˆ«ìë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”")

    return Player(name, gender, job)

# ì‚¬ë§ ì´ìŠˆ
def ask_resurrection():
    while True:
        print("\nğŸ’€ ë‹¤ì‹œ ì‚´ê¸¸ ì›í•˜ë‚˜ ì¸ê°„...?")
        print("1. ë„¤")
        print("2. í•„ìš”ì—†ì–´.")
        
        choice = input("\nì„ íƒí•´ì£¼ì„¸ìš” (1 ë˜ëŠ” 2): ").strip()
        
        if choice == "1":
            print("\nâœ¨ ë‹¹ì‹ ì€ ì‹ ë¹„í•œ í˜ìœ¼ë¡œ ë¶€í™œí–ˆìŠµë‹ˆë‹¤")
            return True
        elif choice == "2":
            print("\nğŸ‘‹ ì¢‹ì€ ì¸ìƒì´ì—ˆë‹¤")
            return False
        else:
            print("\nâŒ 1 ë˜ëŠ” 2ë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”!")

# ë©”ì¸ ê²Œì„ ì‹¤í–‰ í•¨ìˆ˜
def main():
    print("ğŸ® RPG ê²Œì„ì„ ì‹œì‘í•©ë‹ˆë‹¤!")
    
    # ìºë¦­í„° ìƒì„±
    player = create_character()
    print("\nìºë¦­í„°ê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
    player.show_status()
    
    # ëª¬ìŠ¤í„° ì‚¬ì „ ì •ì˜
    monster_dict = {'ìŠ¬ë¼ì„': 1, 'ê³ ë¸”ë¦°': 2, 'ì˜¤í¬': 3}
    
    # ê²Œì„ ì§„í–‰
    while True:
        # ëœë¤í•˜ê²Œ ëª¬ìŠ¤í„° ì„ íƒ
        monster_name = random.choice(list(monster_dict.keys()))
        monster_level = monster_dict[monster_name]
        monster = Monster(monster_name, monster_level)
        
        victory = battle(player, monster)
        
        if not victory:
            if ask_resurrection():  # ë¶€í™œ ì„ íƒ
                # ìºë¦­í„° ìƒíƒœ íšŒë³µ
                player.hp = player.max_hp
                if hasattr(player, 'max_mp'):
                    player.mp = player.max_mp
                player.show_status()
                continue  # ê²Œì„ ê³„ì†
            else:
                print('ğŸª¦ ê²Œì„ì˜¤ë²„')
                return
            
        # ë³´ìŠ¤ ëª¬ìŠ¤í„° ëœë¤ ì¶œí˜„ (20% í™•ë¥ )
        if random.random() < 0.2:
            if boss_encounter(player):
                boss = BossMonster('ëŒ€ì™• ê³ ë¸”ë¦°', 3, monster.hp * 2, monster.power * 2)
                if not battle(player, boss):
                    if ask_resurrection():  # ë¶€í™œ ì„ íƒ
                        # ìºë¦­í„° ìƒíƒœ íšŒë³µ
                        player.hp = player.max_hp
                        if hasattr(player, 'max_mp'):
                            player.mp = player.max_mp
                        player.show_status()
                        continue  # ê²Œì„ ê³„ì†
                    else:
                        print('ğŸª¦ ê²Œì„ì˜¤ë²„')
                        return
        
        # ì‚¬ëƒ¥ ê³„ì† ì—¬ë¶€ í™•ì¸
        while True:
            print("\nğŸ¤” ì‚¬ëƒ¥ì„ ê³„ì† í•˜ì‹œê² ìŠµë‹ˆê¹Œ?")
            print("1. âš”ï¸ ê³„ì†í•œë‹¤")
            print("2. ğŸšª ê²Œì„ì„ ì¢…ë£Œí•œë‹¤")
            
            choice = input("\nì„ íƒí•´ì£¼ì„¸ìš” (1 ë˜ëŠ” 2): ").strip()
            
            if choice == "1":
                print("\nğŸ’ª ì‚¬ëƒ¥ì„ ê³„ì†í•©ë‹ˆë‹¤!")
                player.show_status()
                break
            elif choice == "2":
                print("\nğŸ® ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤...")
                print(f"ìµœì¢… ë ˆë²¨: {player.level}")
                return
            else:
                print("\nâŒ 1 ë˜ëŠ” 2ë§Œ ì…ë ¥í•´ì£¼ì„¸ìš”!")

if __name__ == "__main__":
    main()
```
